```python
nodes = [line.rstrip('\n') for line in open('nodes.txt')]
edges = [[item.strip() for item in line.rstrip('\r\n').split('\t')] for line in open('edges.txt')]
```

```python
file = open('edges.txt', 'r')

test = file.read().splitlines()
```

```python
file = open('edges.txt', 'r')
temp=[]
edges=[]
for line in file.read().splitlines():
    for name1, name2 in line.split(' '):
        temp.append(name1, name2)
edges.append(temp)
```

```python
test
```

```python
## Create and process the graph
```

```python
import matplotlib.pylab as plt
import networkx as nx

G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

```python
G.number_of_nodes()
```

```python
G.number_of_edges()
```

```python
S = [G.subgraph(c).copy() for c in nx.connected_components(G)]
```

```python
S=sorted(S, key=len, reverse=True)
```

```python
sub=S[1]
# S[1] should have 21 nodes
# S[0] is the big one that takes forever to plot
len(sub)
```

```python
print(f'There are {sub.number_of_nodes()} nodes, and')
print(f'There are {sub.number_of_edges()} edges in the subgraph')
```

```python
nx.draw(sub, node_size=100, with_labels=True)
```
